```python
# Create Function For Collecting Metrics
def eval_model(name, y_true, y_pred):
    print(f'{name}')
    print('-' * len(name))
    print('Accuracy :', round(accuracy_score(y_true, y_pred), 3))
    print('Precision:', round(precision_score(y_true, y_pred), 3))
    print('Recall   :', round(recall_score(y_true, y_pred), 3))
    print('F1 Score :', round(f1_score(y_true, y_pred), 3))
```

```python
# Create Function To Upsample Minority Class
def upsample_to_ratio(X, y, ratio=1.0, random_state=12345):
    """Upsample Minority To Reach (Minority/Majority) == Ratio. """
    df = pd.concat([X, y.rename('target')], axis=1)
    df_min = df[df['target'] == 1]
    df_maj = df[df['target'] == 0]
    need = int(ratio * len(df_maj)) - len(df_min)
    if need > 0:
        df_min_extra = df_min.sample(n=need, replace=True, random_state=random_state)
        df_new = pd.concat([df_maj, df_min, df_min_extra], axis=0)
    else:
        df_new = pd.concat([df_maj, df_min], axis=0)
    df_new = shuffle(df_new, random_state=random_state)
    return df_new.drop(columns='target'), df_new['target']
# Create Upsample Model
X_up, y_up = upsample_to_ratio(features_train, target_train, ratio=1.0)
lr_up = LogisticRegression(solver='liblinear', random_state=12345)
lr_up.fit(X_up, y_up)
pred_up = lr_up.predict(features_valid)
eval_model('Upsample (ratio=1.0)', target_valid, pred_up)
```

```python
# Create Function To Downsample Majority Class
def downsample_to_ratio(X, y, ratio=1.0, random_state=12345):
    """Downsample Majority So (Minority / Majority) == ratio."""
    df = pd.concat([X, y.rename('target')], axis=1)
    df_min = df[df['target'] == 1]
    df_maj = df[df['target'] == 0]
    keep_maj = int(len(df_min) / ratio)
    keep_maj = min(keep_maj, len(df_maj))
    df_maj_down = df_maj.sample(n=keep_maj, random_state=12345)
    df_new = shuffle(pd.concat([df_maj_down, df_min], axis=0), random_state=random_state)
    return df_new.drop(columns='target'), df_new['target']
# Create Downsample Model
X_down, y_down = downsample_to_ratio(features_train, target_train, ratio=1.0)
lr_down = LogisticRegression(solver='liblinear', random_state=12345)
lr_down.fit(X_down, y_down)
pred_down = lr_down.predict(features_valid)
eval_model('Downsample (ratio=1.0)', target_valid, pred_down)
```

```python
# --- RF tuner ---
from sklearn.model_selection import StratifiedKFold, RandomizedSearchCV
import numpy as np

# Define Tuning Function
def tune_rf(X_train, y_train, scorer="f1_weighted",
            n_iter=60, cv_splits=5, random_state=12345, n_jobs=-1, verbose=1):
    rf = RandomForestClassifier(class_weight="balanced",
                                random_state=random_state, n_jobs=n_jobs)

    param_space = {
        "n_estimators": np.arange(28, 64, 1),
        "max_depth": [8, 9, 10, 11, 12, 13],
        "min_samples_split": [2, 3, 5],
        "min_samples_leaf": [1, 2],
        "max_features": ["sqrt", 0.8],
        "bootstrap": [True],
        "criterion": ["gini", "entropy"],
    }

    cv = StratifiedKFold(n_splits=cv_splits, shuffle=True, random_state=random_state)
    search = RandomizedSearchCV(
        rf, param_space, n_iter=n_iter, scoring=scorer, cv=cv,
        refit=True, random_state=random_state, n_jobs=n_jobs, verbose=verbose,
        return_train_score=False
    )
    search.fit(X_train, y_train)

    best_model = search.best_estimator_
    best_params = search.best_params_
    print("Best RandomForest params:")
    for k, v in best_params.items():
        print(f"  {k}: {v}")
    return best_model, best_params
```

```python
# Apply Tuning
best_rf, best_params = tune_rf(features_train, target_train, scorer="f1_weighted")

# Test Tuned Model
pred_valid = best_rf.predict(features_valid)
eval_model("Random Forest: Tuned", target_valid, pred_valid)
```
