```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

dataset_list = pd.read_csv('/datasets/datasets.csv')
path = '/datasets/' + dataset_list['file'][36]
car_data = pd.read_csv(path)

#print(car_data.shape)

#print(car_data.columns)

#print(car_data.info())

#print(car_data.isna().sum())

#print(car_data.duplicated().sum())

car_data = car_data.drop_duplicates()

#print(car_data.duplicated().sum())

#print(car_data.dtypes)

#numbers = car_data.select_dtypes(include='number')
#objects = car_data.select_dtypes(include='object')

#print(numbers.dtypes)
#print(objects.dtypes)

date_cols = ['DateCrawled', 'DateCreated', 'LastSeen']
for cols in date_cols:
    car_data[cols] = pd.to_datetime(car_data[cols])


#car_data.info()

#car_data['PostalCode'] = car_data['PostalCode'].astype('object')

#car_data.info()

#car_data.isna().sum()


#car_data['Price'].describe()

car_data = car_data[car_data['Price'] != 0]

#car_data['Price'].describe()

#car_data['Price'].sort_values().unique()[:100]

#car_data100 = car_data[car_data['Price'] <= 100]

#print(car_data100.head(20))
#car_data['Price'].describe()

lower_bound = car_data['Price'].quantile(.20)
car_data = car_data[car_data['Price'] >= lower_bound]
#car_data['Price'].describe()

#higher_bound = car_data['Price'].quantile(.75)

#upper_bound = car_data[car_data['Price'] >= higher_bound]

#print(upper_bound['Price'].sort_values())
#twenty = car_data[car_data['Price'] == 20000]
#twenty.shape
#upper_bound.shape

#np.sort(car_data['Price'].unique())[::-1][:200]

#plt.boxplot(car_data['Price'])
#plt.xlabel('Price')
#plt.title('Boxplot Of Car Prices')
#plt.show()
print(car_data.info())
car_data.describe()
```

```python
print(car_data.head(5))

bins = [0, 25_000, 50_000, 75_000, 100_000, 150_000]
labels = [
    '0-25k',
    '25-50k',
    '50-75k',
    '75-100k',
    '100-150k',
]

car_data['Mileage_bin'] = pd.cut(
    car_data['Mileage'],
    bins=bins,
    labels=labels,
    include_lowest=True
)

price_by_mileage = (
    car_data
    .groupby('Mileage_bin')['Price']
    .agg(['count', 'mean', 'median'])
)

#print(price_by_mileage)

#No Values Over 150k
#Value Goes Up As Mileage Goes Down
#Except Between 0-25k, Value Actually Takes A Drop From 9847 To 8342

insights25k = car_data[car_data['Mileage_bin'] == '0-25k']
                        

#print(insights_25k.shape)

#print(insights_25k['Price'].describe())

#insights_25k.corr()

threshold = insights25k['Price'].quantile(.25)
bottom25 = insights25k[insights25k['Price'] <= threshold]

print(bottom25.describe())

print(bottom25['Power'].unique())

toptiercars = bottom25[(bottom25['Power'] >= 800) & (bottom25['Power'] <= 1200)]
print(toptiercars[['Power', 'Brand', 'Model', 'Price']].sort_values(by='Power', ascending=False))

lowertiercars = car_data[(car_data['Power'] < 50) & (car_data['Power'] >= 40)]
print('Cars Between 40 & 50 HP')
print(lowertiercars[['Power', 'Brand', 'Model', 'RegistrationYear']])

car_data = car_data[(car_data['Power'] <= 800) & (car_data['Power'] >= 50)]

print(price_by_mileage)
```

```python
car_data.describe()

#Replace 'Power' Column To 'PowerHP'
#Replace 'RegistrationYear' To 'YearOfModel'
car_data = car_data.rename(columns={
    'RegistrationYear': 'YearOfModel',
    'Power': 'PowerHP'
})

car_data.columns

car_data = car_data[(car_data['YearOfModel'] <= 2026) & (car_data['YearOfModel'] >= 1950 )]

car_data = car_data[car_data['RegistrationMonth'] != 0 ]

lowerhp = car_data[(car_data['PowerHP'] >= 50) & (car_data['PowerHP'] <= 150)]

car_data = car_data.drop(columns=['NumberOfPictures'])

print(lowerhp[['Brand', 'Model', 'PowerHP', 'YearOfModel']].sort_values(by='PowerHP', ascending=True))

print(car_data.describe())
```
