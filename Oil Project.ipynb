````python
# Split, Train And Test The Model For Each Region
def train_and_evaluate_region(data, region_name):
    features = data[['f0', 'f1', 'f2']]
    target = data['product']
    features_train, features_valid, target_train, target_valid = train_test_split(features, target, test_size=0.25, random_state=12345)

    model = LinearRegression()
    model.fit(features_train, target_train)

    preds = model.predict(features_valid)

    #Calculate RMSE And Average Predicted Volume
    rmse = mean_squared_error(target_valid, preds, squared=False)
    avg_pred = preds.mean()
    avg_actual = target_valid.mean()

    print(f'üìç {region_name}')
    print(f'RMSE: {rmse:.3f}')
    print(f'Average Predicted Reserves: {avg_pred:.3f}')
    print(f'Average actual reserves: {avg_actual:.3f}')
    print('-' * 45)

    return model, features_valid, target_valid, preds

#Run Models For Each Region
model_0, fv0, tv0, pred0 = train_and_evaluate_region(geo0, 'Region 0')
model_1, fv1, tv1, pred1 = train_and_evaluate_region(geo1, 'Region 1')
model_2, fv2, tv2, pred2 = train_and_evaluate_region(geo2, 'Region 2')
````

````python
# Profit from model predictions

#Create Function To Predict Profits From Regions
def profit_from_predictions(target_valid, target_pred, n_select=top_wells, price=price_per_unit, budget=Budget):
    target_pred_s = pd.Series(target_pred, index=target_valid.index)
    top_idx = target_pred_s.sort_values(ascending=False).index[:n_select]

    total_selected_reserves = target_valid.loc[top_idx].sum()
    revenue = price * total_selected_reserves
    profit = revenue - budget

    return profit, revenue, total_selected_reserves, top_idx
````

````python
def eval_region_revenue(model, name, target_valid, preds):

    profit, revenue, total_res, top_idx = profit_from_predictions(target_valid, preds)

    avg_pred = pd.Series(preds, index=target_valid.index).loc[top_idx].mean()
    avg_actual = target_valid.loc[top_idx].mean()

    print(f'üìç {name}')
    print(f'Selected Wells: {len(top_idx)}')
    print(f'Total Reserves (Actual) Of Selected Wells: {total_res:,.2f} Thousand Barrels')
    print(f'Revenue (@ $4,500 Per Unit): ${revenue:,.0f}')
    print(f'Profit After $100M Budget: ${profit:.0f}')
    print(f'Avg Predicted (Selected): {avg_pred:.2f} | Avg Actual (Selected): {avg_actual:.2f}')
    print('-'*60)

eval_region_revenue(model_0, 'Region 0', tv0, pred0)
eval_region_revenue(model_1, 'Region 1', tv1, pred1)
eval_region_revenue(model_2, 'Region 2', tv2, pred2)
````

````python
#Bootstrapping profit, CI, and risk of loss

#Create Validation Dataframe With Preds & Actuals
def create_valid_df(model, features_valid, target_valid, preds):
    valid_df = features_valid.copy()
    valid_df['pred'] = preds
    valid_df['actual'] = target_valid.values
    return valid_df

valid_df0 = create_valid_df(model_0, fv0, tv0, pred0)
valid_df1 = create_valid_df(model_1, fv1, tv1, pred1)
valid_df2 = create_valid_df(model_2, fv2, tv2, pred2)

#Choose Top 200 Wells By Predicted, Compute Profit Using Actual Reserves
def profit_from_sample(sample_df):
    top = sample_df.sort_values('pred', ascending=False).head(top_wells)
    revenue = top['actual'].sum() * price_per_unit
    profit = revenue - Budget
    return profit

def bootstrap_profit(valid_df, n_boot=sample_size, study_size=Study_size, seed=Random_state, return_samples=False):
    rng = np.random.default_rng(seed)
    profits = np.empty(n_boot, dtype=float)
    n = len(valid_df)
    # If Validation Set < Study Size, Sample With Replacement Anyway
    for b in range(n_boot):
        idx = rng.integers(0, n, size=study_size)
        sample = valid_df.iloc[idx]
        profits[b] = profit_from_sample(sample)
    mean_profit = profits.mean()
    ci_low, ci_high = np.percentile(profits, [2.5, 97.5])
    risk_of_loss = (profits < 0).mean() #Probability Of Negative Profit
    out = {
        'mean_profit': mean_profit,
        'ci_low': ci_low,
        'ci_high': ci_high,
        'risk_of_loss': risk_of_loss,
        'profits': profits
    }
    if return_samples:
        out['profits'] = profits
    return out
def run_region(valid_df, name):
    stats = bootstrap_profit(valid_df, n_boot=sample_size, study_size=Study_size, seed=Random_state)
    print(f'üìç {name}')
    print(f'Mean Profit: ${stats["mean_profit"]:,.0f}')
    print(f'95% CI: [${stats["ci_low"]:,.0f}, ${stats["ci_high"]:,.0f}]')
    print(f'Risk Of Loss (<$0): {stats["risk_of_loss"]*100:.2f}%')
    print('-'*60)
    return stats

#Run All Three Regions (geo0, geo1, geo2)
stats0 = run_region(valid_df0, 'Region 0')
stats1 = run_region(valid_df1, 'Region 1')
stats2 = run_region(valid_df2, 'Region 2')

#Stop Samples From Printing
pretty0 = {k: v for k, v in stats0.items() if k != 'profits'}
pretty1 = {k: v for k, v in stats1.items() if k != 'profits'}
pretty2 = {k: v for k, v in stats2.items() if k != 'profits'}

print(pretty0)
print(pretty1)
print(pretty2)
````
